```python
# Howard Wolowitz has a cylindrical tank in which he has to fill water in
# certain hours. For every one hour, his house will receive N liters of water
# from the corporation. Find out whether he will be able to fill the tank with
# water in X hours. The radius and height of the cylinder are given as the
# input.
# Input Format:
# Input consists of 4 integer values.
# The first input corresponds to the radius of the tank.
# The second input corresponds to the height of the tank.
# The third input corresponds to the amount of water the tank receives every
# hour.
# The fourth input corresponds to the total number of hours the tank receives
# water.
# Question 5
# Output Format:
# The output should be any one of the following strings.
# "The tank can be filled within x hours" or "The tank cannot be filled within x
# hours".
# Sample Input 1:
# 5
# 7
# 100
# 6
# Sample Output 1:
# The tank can be filled within 6 hours.
# Question 5
# Explanation:
# Here, the volume of the tank is less than or equal to the total amount of
# water and hence the output is "The tank can be filled within 6 hours".
# Sample Input 2:
# 8
# 5
# 50
# 3
# Sample Output 2:
# The tank cannot be filled within 3 hours
# Explanation:
# Here, the volume of the tank is greater than the total amount of water and
# hence the output is"The tank cannot be filled within 3 hours"
radius = float(input())
height = int(input())
waterrecieve = int(input())
hrs = int(input())
area = 3.14 * (radius ** 2) * height
total_water = waterrecieve * hrs 
if total_water >= area:
    print("the tank can be fill within3hrs")
else:
    print("the tank cannt be fill within 3hrs")
```

```python
# Question 6
# Sansa and Daenerys are close friends. They both are studying in the same
# school. Now they are on their summer vacation. As they are bored, they ask
# their parents to take them to an exhibition. There Sansa and Daenerys play
# a game. In this game, there are three boxes with some numbers written on
# their top. There is a treasure in one of the three boxes and the treasure is
# present in the box with the second largest number on its top. Also, to open
# the box, they need to decode the correct code of this box. The clue given to
# them to find the code is that it is the largest number that divides all the
# three given numbers. So, now help Sansa and Daenerys to decode the
# code.
# Question 6
# Input Format:
# Input consists of three integers.
# The first input corresponds to the number of the first box.
# The second input corresponds to the number of the second box.
# The third input corresponds to the number of the third box.
# Output Format:
# The output consists of two integers along with a string, one is a box
# number in which treasure is present and the next one is a code to open the
# box.
# Question 6
# Sample Input 1:
# 2
# 4
# 6
# Sample Output 1:
# The treasure is in the box which has the number 4.
# The code to open the box is 2.
# Explanation:
# Here the second largest number is 4 and the largest number which divides
# all the 3 numbers is 2 hence the output is "The treasure is in the box which
# has the number 4.
# The code to open the box is 2."
# Question 6
# Sample Input 2:
# 7
# 5
# 6
import math

num1 = int(input())
num2 = int(input())
num3 = int(input())

number = [num1, num2, num3]
number.sort()

second_largest = number[1]


gcd1 = math.gcd(math.gcd(num1, num2), num3)

print(f"The treasure is in the box which has the number {second_largest}.")
print(f"The code to open the box is {gcd1}.")
```
